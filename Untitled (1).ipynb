### Рекомендации

```python
users_pivot=df.pivot_table(index=["userId"],columns=["title"],values="rating")
users_pivot.fillna(0,inplace=True)
users_pivot.head()
```

```python
from scipy.sparse import csr_matrix
import numpy as np
from sklearn.neighbors import NearestNeighbors
import pandas as pd

films_df_matrix = csr_matrix(users_pivot.values)
model_knn = NearestNeighbors(metric='cosine', algorithm='brute')
model_knn.fit(films_df_matrix)

def find_favorite_films(User_id, count=10):
    user_index = users_pivot.index.get_loc(User_id)
    
    distances, indices = model_knn.kneighbors(films_df_matrix[user_index], n_neighbors=count+1)
    
    favorite_indices = indices[0][1:]
    
    favorite_films=pd.DataFrame({"favorite films ":[users_pivot.columns[idx] for idx in favorite_indices]})
    return favorite_films
```

```python
find_favorite_films(1)
```

```python
avg_ratings = df.groupby('title')['rating'].mean().reset_index().rename(columns={'rating': 'avg_rating'})

avg=pd.DataFrame(avg_ratings).sort_values('avg_rating',ascending=False)
avg.head()
```

```python
cnt_ratings = df.groupby('title')['rating'].count().reset_index().rename(columns={'rating': 'count_rating'})

cnt=pd.DataFrame(cnt_ratings).sort_values('count_rating',ascending=False)
cnt.head()
```

```python
popularite=avg.merge(cnt,on='title')
popularite.head()
```

```python
m= popularite['count_rating'].quantile(0.9)
C= popularite['avg_rating'].mean()

def weighted_rating(row):
    v = row['count_rating']
    R = row['avg_rating']
    return (v/(v+m) * R) + (m/(m+v) * C)

popularite['score'] = popularite.apply(weighted_rating,axis=1)
```

```python
u_df = df.drop_duplicates(subset=['title'])

def getGenres(row):
    return " ".join(u_df[u_df['title']==row['title']]['genres'].iloc[0].split("|"))

popularite['genres'] = popularite.apply(getGenres,axis=1)
```

### Рекомендации по контенту

```python
def createOverview(row):
    return row['title_wy'] + " " + row['genres'].replace("|"," ")

films['overview'] = films.apply(createOverview,axis=1)

films.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(stop_words='english')
films['overview'] = films['overview'].fillna('')
tfidf_matrix = tfidf.fit_transform(films['overview'])
```

```python
tfidf_matrix.shape
```

```python
from sklearn.metrics.pairwise import linear_kernel

cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
cosine_sim
```

```python
indices = pd.Series(films.index, index=films['title_wy'])
indices
```

```python
def get_recommendations(title, cosine_sim=cosine_sim):
    try:
        idx = indices[title.lower()]
    except:
        titles = indices.index.tolist()
        for i in range(0, len(titles)):
            val = titles[i]
            if title.lower() in val.lower():
                idx = i
                break


    sim_scores = list(enumerate(cosine_sim[idx]))

    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    sim_scores = sim_scores[1:11]

    movie_indices = [i[0] for i in sim_scores]

    return films['title'].iloc[movie_indices]
```

```python
get_recommendations('Toy Story').values
```
